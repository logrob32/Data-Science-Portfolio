```python
import pandas as pd
```

```python
states = pd.read_csv("uscounties_clustered.csv")
```

```python
states.head()
```

```python
states.groupby('Cluster')['Population'].sum().describe()
```

```python
states['Democrat'] = 0
states['Republican'] = 0
states['Other'] = 0
states.head()
```

```python
votes = pd.read_csv("countypres_2000-2020.csv")
```

```python
votes.head()
```

```python
votes_2020 = votes.loc[votes['year']==2020].copy()
votes_2020.head()
```

```python
votes_2020[votes_2020['county_fips'].isnull()]
```

```python
votes_2020.loc[votes_2020['county_name']=="DISTRICT OF COLUMBIA", "county_fips"] = 11001
votes_2020.dropna(subset=["county_fips"], inplace=True)
votes_2020[votes_2020['county_fips'].isnull()]
```

```python
votes_2020['county_fips'] = votes_2020['county_fips'].astype('int')
```

```python
votes_2020.head()
```

```python
votes_2020['candidate'].unique()
```

```python
votes_2020.loc[~votes_2020['party'].isin(["DEMOCRAT", "REPUBLICAN"]), 'candidatevotes']
```

```python
for fip in states['County Fips']:
    fip_votes = votes_2020.loc[votes_2020['county_fips']==fip]
    states.loc[states['County Fips']==fip, 'Democrat'] = fip_votes.loc[fip_votes['party']=="DEMOCRAT", 'candidatevotes'].sum()
    states.loc[states['County Fips']==fip, 'Republican'] = fip_votes.loc[fip_votes['party']=="REPUBLICAN", 'candidatevotes'].sum()
    states.loc[states['County Fips']==fip, 'Other'] = fip_votes.loc[~fip_votes['party'].isin(["DEMOCRAT", "REPUBLICAN"]), 'candidatevotes'].sum()
```

```python
states['total_votes'] = states['Democrat'] + states['Republican'] + states['Other']
states.tail()
```

```python
states[states['total_votes']==0]
```

```python
states['percent_dem'] = states['Democrat']/states['total_votes']
states['percent_rep'] = states['Republican']/states['total_votes']
states['dem_lead'] = states['percent_dem']-states['percent_rep']
```

```python
states.tail()
```

```python
states.to_csv("county_votes.csv")
```

```python
states_clustered = states.groupby('Cluster').sum()
```

```python
states_clustered['percent_dem'] = states_clustered['Democrat']/states_clustered['total_votes']
states_clustered['percent_rep'] = states_clustered['Republican']/states_clustered['total_votes']
states_clustered['dem_lead'] = states_clustered['percent_dem']-states_clustered['percent_rep']
```

```python
states_clustered = states_clustered.reset_index()
states_clustered.head()
```

```python
def vote_share_state(cluster):
    dem_lead = states_clustered.loc[states_clustered['Cluster']==cluster, 'dem_lead'].mode()
    return dem_lead
```

```python
states['state_vote_share'] = states['Cluster'].apply(lambda x: vote_share_state(x))
states.tail()
```

```python
states.to_csv("state_votes.csv")
```

```python
states_clustered['winner'] = states_clustered['dem_lead'].apply(lambda x: "Democrat" if x > 0 else "Republican")
```
